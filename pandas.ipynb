```python
from typing import Tuple
from sklearn.datasets import load_iris
import sklearn.model_selection
import pandas as pd # Import pandas for DataFrame and Series

def cargar_datos() -> Tuple[pd.DataFrame, pd.Series, pd.DataFrame, pd.Series]:
    iris_dataset = load_iris()
    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
        iris_dataset['data'], iris_dataset['target'], random_state=0, test_size=0.1
    )

    # Clean column names to remove spaces and special characters for robust access
    cleaned_feature_names = [
        name.replace(' (cm)', '_cm').replace(' ', '_') for name in iris_dataset['feature_names']
    ]

    return (pd.DataFrame(X_train, columns=cleaned_feature_names),
            pd.Series(y_train, name='target'),
            pd.DataFrame(X_test, columns=cleaned_feature_names),
            pd.Series(y_test, name='target'))

def predict(X: pd.DataFrame) -> list:
  y_predicted = []
  for x in X.itertuples():
    try:
      y_i = predict_instance(x)
    except NameError:
      raise NameError("Definir la funcion predict_instance")
    y_predicted.append(y_i)
  return y_predicted

def accuracy(y_predicted: list, y_real: list) -> float:
  TP_TN = sum([y_i == y_j for (y_i, y_j) in zip(y_predicted, y_real)])
  P_N = len(y_real)
  return TP_TN / P_N
```

```python
X_train, y_train, X_test, Y_test = cargar_datos()
X_train.head()
```

```python
def predict_instance(x) -> int:
    """
    Predice la clase para instancia x.
    """
    # Use the cleaned column name 'petal_width_cm'
    if x._asdict()['petal_width_cm'] < 0.8:
        prediction = 0  # Iris-setosa
    elif (x._asdict()['petal_width_cm'] > 1) and (x._asdict()['petal_width_cm'] < 1.7):
        prediction = 1  # Iris-versicolor
    else:
        # Placeholder for other classes, will be refined later
        prediction = 2 # Default to Clase 2 for now

    return prediction
```

```python
y_pred = predict(X_train)
# Completar el algoritmo para mejorar esta performance.

print(f"Accuracy: {round(accuracy(y_pred, y_train), 3)}")
```

# 2. Dividir Datos en Entrenamiento y Prueba:
Se utiliza train_test_split para separar 'X' y 'y' en conjuntos de entrenamiento (X_train, y_train) y conjuntos de prueba (X_test, y_test). Esto asegura que el modelo se entrene con un subconjunto de datos y se evalúe con datos no vistos.

```python
import sklearn.model_selection
X = cancer_dataset.drop(columns=['target'])
y = cancer_dataset['target']
#TEST_SIZE = 0.2

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=4, test_size=0.2)
```

# 3. Inicializar el Modelo de Árbol de Decisión:
Se crea una instancia del DecisionTreeClassifier con parámetros específicos, como max_depth y criterion, que definen la complejidad y la estrategia de división del árbol.

```python
from sklearn.tree import DecisionTreeClassifier
arbol = DecisionTreeClassifier(max_depth=1, criterion="entropy")
```

# 4. Entrenar el Modelo:
Se entrena el árbol de decisión utilizando el método fit() con los datos de entrenamiento (X_train, y_train). Durante este proceso, el modelo aprende los patrones y relaciones en los datos para realizar predicciones posteriores.

```python
arbol.fit(X_train, y_train)
```

# 5. Realizar Predicciones en el Conjunto de Prueba:
Una vez entrenado, el modelo se utiliza para hacer predicciones (y_pred) sobre el conjunto de prueba (X_test), que contiene datos que el modelo nunca ha visto antes.

```python
from sklearn.metrics import classification_report

y_pred = arbol.predict(X_test)
print("Reporte de Clasificación en el conjunto de prueba:")
print(classification_report(y_test, y_pred))
```

# 6. Evaluar el Modelo con Métricas de Rendimiento:
Se calculan métricas de evaluación como accuracy (exactitud) y la matriz de confusión para evaluar el rendimiento del modelo en el conjunto de prueba y entrenamiento.

```python
print(f"Predicciones:   {y_pred}\nValores reales: {y_test.to_numpy()}")
# Podemos calcular el accuracy (exactitud) comparando los valores predichos contra los reales,
# para ello contamos cuántas coincidencias hubo y dividimos por la cantidad de comparaciones que hicimos:
print(f"Accuracy sobre el test set: {np.mean(y_pred == y_test)}")

#También podemos invocar al método score que viene con los DecisionTreeClassifier
print(f"Accuracy sobre el test set: {arbol.score(X_test, y_test)}")

print(f"Score sobre el training set: {arbol.score(X_train, y_train):.2f} ¿qué indica este número?")
```

```python
print("Matriz de confusión:")
confusion = sklearn.metrics.confusion_matrix(y_pred=y_pred, y_true=y_test)
display(pd.DataFrame(confusion, columns=["0", "1"], index=["0", "1"]))
```
